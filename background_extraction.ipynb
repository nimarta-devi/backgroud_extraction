```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

video = cv2.VideoCapture('highway.mp4')

FOI = video.get(cv2.CAP_PROP_FRAME_COUNT) * np.random.uniform(size=30)

#creating an array of frames from frames chosen above
frames = []
for frameOI in FOI:
    #print(cv2.CAP_PROP_POS_FRAMES)
    video.set(cv2.CAP_PROP_POS_FRAMES, frameOI)
    ret, frame = video.read()
    #cv2_imshow(frame)
    frames.append(frame)
```

```
#calculate the average
backgroundFrame_mean = np.mean(frames, axis=0).astype(dtype=np.uint8)    
cv2.imwrite("bg.jpg",backgroundFrame_mean)
cv2_imshow(backgroundFrame_mean)
```

```
#calculate the average
backgroundFrame_median = np.median(frames, axis=0).astype(dtype=np.uint8)    
cv2.imwrite("bg.jpg",backgroundFrame_median)
cv2_imshow(backgroundFrame_median)
```
